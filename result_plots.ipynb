```python
import matplotlib.pyplot as plt
```

```python
Version = ['Both', 'ROS2', 'ROS1', 'Not\nspecified']
Quantity = [2, 2, 11, 24]

plt.barh(Version, Quantity, color = 'darkgray', edgecolor = 'black')

for index, value in enumerate(Quantity):
    plt.text(value, index, "  " + str(value))

plt.xlim([0, 30])
#plt.show()
plt.savefig('figures/ros_version.pdf', bbox_inches='tight')
```
